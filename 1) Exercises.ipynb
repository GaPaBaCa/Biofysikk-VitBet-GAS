**The diffusion equation**  

## **Exercise 1a)**  
Show that  
$$
\phi(x,t) = \frac{1}{\sqrt{4\pi D t}} \int_{-\infty}^{\infty} dy\ h(y) e^{-\frac{(x-y)^2}{4Dt}}
$$


is a solution of the **diffusion equation**.

Let $\phi(x,t)$ represent a distribution of particles along the $x$-axis at time $t \geq 0$. Give a physical interpretation of $h(y)$.  
*Hint: let $t \to 0$ in Eq. (6).*

---


**Solution** : We consider the function $\phi(x,t)$ which describes the spatial distribution of particles along the $x$ -axis at a given time $t$. Our objective is to demonstrate that this function satisfies **diffusion equation**, which governs the temporal and spatial evolution of diffusing quantities such as heat or particles. 

The **diffusion equation** is given by:

$$
\frac{\partial \phi(x,t)}{\partial t} = D \frac{\partial^2 \phi(x,t)}{\partial x^2}.
$$

This equation states the rate of change of $\phi(x,t)$ with respect to time is proportional to the second spatial derivative of $\frac{\partial^2 \phi(x,t)}{\partial x^2}$. with respect to the particles(x), indictaing that diffusion is driven by spatial gradients in the distribution.

To proceed with the proof, we decompose $\phi(x,t)$ into two functions f(x) and g(x),
$\phi(x,t)= f(t)g(x,t)$,
 where 
$$f(t)= \frac{1}{\sqrt{4\pi D t}}$$
and
$$g(x,t) = \int_{-\infty}^{\infty} dy\ h(y) e^{-\frac{(x-y)^2}{4Dt}}$$




**Computation of Partial Derivatives**: 

To proceed, we compute the partial derivatives with respect to $t$ and $x$, respectively. The variables involved are:  
- $x$, $y$, $t$, $D$

We aim to determine:
 $$\frac{\partial}{\partial t}[f(t) g(t)] \quad and \quad \frac{\partial^2 \phi(x,t)}{\partial x^2}$$


 **Derivative of $f(t)$ with respect to $t$**

Starting with $\frac{\partial f}{\partial t}$. The function $f(t)$ can be written as $(4\pi D t)^{-\frac{1}{2}}$
Therefore, we could say we are trying to find:
 $$\frac{\partial }{\partial t}[(4\pi D t)^{-\frac{1}{2}}]$$
Applying the chain rule:
 $${-\frac{1}{2}} (4\pi D t)^{-\frac{3}{2}} 4 \pi D$$
 $${-\frac{4 \pi D}{2 (4\pi D)^{\frac{1}{2}} t^{\frac{3}{2}}}}$$
 $${-\frac{1}{4\sqrt{\pi D}t^{\frac{3}{2}}}}$$
 Thus, we have determined the partial derivative of $f(x)$ with respect to $t$: 
$$\frac{\partial f}{\partial t} = {-\frac{1}{4\sqrt{\pi D}t^{\frac{3}{2}}}}$$
$$\frac{\partial f}{\partial t} = {{-\frac{1}{2 t}} \frac{1}{\sqrt{4 \pi D t}}}$$

Next, we compute the partial derivative of $g(x,t)$ with respect to $t$: 
 $$\frac{\partial g}{\partial t}$$ 
 Since the function $g(x,t)$ contains an exponential term, we apply the chain rule to differentiate the exponent: 
 $$\frac{\partial }{\partial t}[e^{-\frac{(x-y)^2}{4Dt}}]$$
Given that the derivative of the exponential function is itelf multiplied by the derivative of its own exponent, we obtain:  
$$\frac{\partial }{\partial t}\left[{-\frac{(x-y)^2}{4Dt}}\right] = {\frac{(x-y)^2}{4Dt^2}}$$
Thus, differentiating the exponential function gives:
$$\frac{\partial }{\partial t}[e^{-\frac{(x-y)^2}{4Dt}}] = e^{-\frac{(x-y)^2}{4Dt}} {\frac{(x-y)^2}{4Dt^2}}$$

Previously, we introduced the decomposition of $\phi(x,t)$ into two functions $f(x)$ and $g(x,t)$. Now, that we have computed their respective derivatives: $$\frac{\partial f}{\partial t} and \frac{\partial g}{\partial t}$$ 
We can proceed to differentiate $\phi(x,t)$. Since $\phi(x,t)$ is defined as the product of $f(x)$ and $g(x,t)$, we apply the product rule:
$$\frac{\partial \phi(x,t)}{\partial t} = \frac{\partial}{\partial t} [f(t) g(t)]$$ 
 Thus, it is necessary to apply the product rule, which states: 
$$\frac{\partial}{\partial t} [f(t) g(t)] = f'(t) g(t) + f(t) g'(t)$$
Rewriting this in terms of $\phi(x,t)$, we obtain:
$$\frac{\partial \phi(x,t)}{\partial t} = f'(t) g(t) + f(t) g'(t)$$ 
We can now express the time derivative of $\phi(x,t)$ as: 
$$\frac{\partial \phi(x,t)}{\partial t} = {{-\frac{1}{2 t}} \frac{1}{\sqrt{4 \pi D t}}} \int_{-\infty}^{\infty} \ h(y) e^{-\frac{(x-y)^2}{4Dt}}dy  + \frac{1}{\sqrt{4\pi D t}} \int_{-\infty}^{\infty} \ h(y) e^{-\frac{(x-y)^2}{4Dt}} {\frac{(x-y)^2}{4Dt^2}}dy$$

**Computation of the Second Partial Derivative**

We now proceed to compute the second partial derivative of $\phi(x,t)$ with respect to $x$: 
$$\frac{\partial^2 \phi(x,t)}{\partial x^2}$$

Since the function $f(x)$ does not contain $x$, it will act as a constant, the function $g(x)$ has $x$ in its exponent, requiring the application of the chain rule. As we need to differentiate twice, we first compute the first derivative.

**First Partial Derivative**

Applying the chain rule, we differentiate the exponent:
$$ \frac{\partial }{\partial x} \left[{-\frac{(x-y)^2}{4Dt}}\right]$$
which simplifies as follows:
 $$\frac{\partial }{\partial x} \left[ -\frac{(x-y)^2}{4Dt} \right] = -\frac{2(x-y)}{4Dt}$$
$$\frac{\partial }{\partial x} \left[ -\frac{(x-y)^2}{4Dt} \right] = -\frac{x-y}{2Dt}$$
Thus, differentiating the exponential function yields:
$$\frac{\partial }{\partial x} [e^{-\frac{(x-y)^2}{4Dt}}] = e^{-\frac{(x-y)^2}{4Dt}} \left(-\frac{x-y}{2Dt}\right)$$
Therefore, the first derivative of $\phi(x,t)$ is:
$$\frac{\partial \phi(x,t)}{\partial x} = \frac{1}{\sqrt{4\pi D t}} \int_{-\infty}^{\infty} \ h(y) e^{-\frac{(x-y)^2}{4Dt}}\left(-\frac{x-y}{2Dt}\right)dy$$
We now proceed to compute the second derivative. 
Since the given expression consists of a product of two functions, it is necessary to apply the product rule. We define the function as follows:
$$ h(x)= e^{-\frac{(x-y)^2}{4Dt}} \quad and \quad  i(x) = \left(-\frac{x-y}{2Dt}\right)$$

 It would then look like this: 
$$\frac{\partial^2 \phi(x,t)}{\partial x^2} \left[e^{-\frac{(x-y)^2}{4Dt}}\left(-\frac{x-y}{2Dt}\right)\right]= \frac{\partial }{\partial x} [h(x) i(x)]$$
Taking the first derivative of these functions: 
$$\frac{\partial}{\partial x}h(x)= e^{-\frac{(x-y)^2}{4Dt}} \left(-\frac{x-y}{2Dt}\right)$$
$$\frac{\partial}{\partial x} i(x)= \left(-\frac{1}{2Dt}\right)$$
Applying the product rule, which states: 
$$\frac{\partial }{\partial x} [h(x) i(x)] = h'(x) i(x) + h(x) i'(x)$$
Substituting the computed derivatives: 
$$\frac{\partial }{\partial x} [h(x) i(x)]= e^{-\frac{(x-y)^2}{4Dt}} \left(-\frac{x-y}{2Dt}\right) \left(-\frac{x-y}{2Dt}\right) + e^{-\frac{(x-y)^2}{4Dt}}\left(-\frac{1}{2Dt}\right)$$
$$\frac{\partial }{\partial x} [h(x) i(x)] = \frac{(x-y)^2}{4D^2t^2} e^{-\frac{(x-y)^2}{4Dt}} -\left(\frac{1}{2Dt}\right)e^{-\frac{(x-y)^2}{4Dt}}$$

**Factoriztion of the Diffusion Equation**

Thus, we can express the equation as follows: 
$$\frac{\partial^2 \phi(x,t)}{\partial x^2} = \frac{1}{\sqrt{4\pi D t}} \frac{(x-y)^2}{4D^2t^2}\int_{-\infty}^{\infty} \ h(y) e^{-\frac{(x-y)^2}{4Dt}}dy + -\left(\frac{1}{2Dt}\right)\int_{-\infty}^{\infty} \ h(y)e^{-\frac{(x-y)^2}{4Dt}}dy$$

To simplify, we factor out D from the denominator and rewrite the intergral terms, yielding: 
$$D\frac{\partial^2 \phi(x,t)}{\partial x^2}= \frac{1}{\sqrt{4\pi D t}} \int_{-\infty}^{\infty} \ h(y) e^{-\frac{(x-y)^2}{4Dt}}\frac{(x-y)^2}{4Dt^2}dy + \int_{-\infty}^{\infty} \ h(y)e^{-\frac{(x-y)^2}{4Dt}}-\frac{1}{2t}dy$$
Further factorization allows us to express the equation as: 
 $$D\frac{\partial^2 \phi(x,t)}{\partial x^2}= \frac{1}{\sqrt{4\pi D t}} \int_{-\infty}^{\infty} \ h(y) e^{-\frac{(x-y)^2}{4Dt}} \left [\frac{(x-y)^2}{4Dt^2} -\frac{1}{2t}\right]dy $$

 **Final Simplification and Conclusion**

 Returning to our previously derived expressions, we now simplify the equation as follows:
$$\frac{\partial \phi(x,t)}{\partial t} = {{-\frac{1}{2 t}} \frac{1}{\sqrt{4 \pi D t}}} \int_{-\infty}^{\infty} \ h(y) e^{-\frac{(x-y)^2}{4Dt}}dy  + \frac{1}{\sqrt{4\pi D t}} \int_{-\infty}^{\infty} \ h(y) e^{-\frac{(x-y)^2}{4Dt}} {\frac{(x-y)^2}{4Dt^2}}dy$$
Through further simplifications, we obtain: 
 $$\frac{\partial \phi(x,t)}{\partial t}= \frac{1}{\sqrt{4 \pi D t}} \int_{-\infty}^{\infty} \ h(y) e^{-\frac{(x-y)^2}{4Dt}}{{-\frac{1}{2 t}}}dy + \frac{1}{\sqrt{4\pi D t}} \int_{-\infty}^{\infty} \ h(y) e^{-\frac{(x-y)^2}{4Dt}} {\frac{(x-y)^2}{4Dt^2}}dy$$
 $$\frac{\partial \phi(x,t)}{\partial t}= \frac{1}{\sqrt{4\pi D t}} \int_{-\infty}^{\infty} \ h(y) e^{-\frac{(x-y)^2}{4Dt}} \left[\frac{(x-y)^2}{4Dt^2} -\frac{1}{2t}\right]dy $$

Recognizing the structure of the diffusion equation, we conclude:
 $$\frac{\partial \phi(x,t)}{\partial t} = D \frac{\partial^2 \phi(x,t)}{\partial x^2}$$
 Thus, we have successfully demonstrated that $\phi(x,t)$  satisfies the diffusion equation. 

 **Proof complete**
